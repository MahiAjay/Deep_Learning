Data Preprocessing


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder 
# takes action on the metadata
from sklearn.compose import ColumnTransformer
#column transformer basically does indexing on the column whiich is metadata
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
url = "https://raw.githubusercontent.com/redashu/Datasets/master/bank.csv" 
df = pd.read_csv(url)
wf = df.copy()
wf
```

```python
wf.info()
```

```python
features = wf.iloc[:,3:-1].values
labels = wf.iloc[:,-1].values
```

```python

pd.DataFrame(features)
```

```python
le = LabelEncoder()
```

```python
features[:,2] = le.fit_transform(features[:,2])
features
```

```python
#preparing metadata
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
#                              encoding type,   method, column ,only numbers before decimal
```

```python
features=np.array(ct.fit_transform(features))
features
```

```python
X_train,x_test,Y_train,y_test = train_test_split(features,labels,test_size=0.2)
```

```python
# feature scaling is a must process for deep learning
SS = StandardScaler()
trained = SS.fit_transform(X_train)
print(trained)
```

Creating ANN Model 

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras import layers
```

```python
ann2= Sequential()
```

```python
#adding intermediate / hidden layers
ann2.add(layers.Dense(20,activation='relu')), # first  hidden layer
ann2.add(layers.Dense(15,activation='relu')),
ann2.add(layers.Dense(10,activation='relu')),
ann2.add(layers.Dense(1,activation='sigmoid'))# output layer
```

```python
#lets compile ANN layers with cost /loss function
ann2.compile(optimizer = 'adam', loss='binary_crossentropy',metrics=['accuracy'])
```
